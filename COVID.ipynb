```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
px
```

```python
base_url = "https://github.com/CSSEGISandData/COVID-19/blob/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv/?raw=true"
confirmed_df = pd.read_csv(base_url)
```

```python
confirmed_df
```

```python
confirmed_df = confirmed_df.drop(columns=['Lat', 'Long', 'Province/State'])
```

```python
confirmed_df
```

```python
confirmed_df = confirmed_df.groupby(by='Country/Region').aggregate(np.sum).T
```

```python
confirmed_df
```

```python
confirmed_df.index.name = "DATE"
```

```python
us_confirmed_df = confirmed_df[['US']]
```

```python
us_confirmed_df
```

```python
us_confirmed_df.columns=['Confirmed']
```

```python
us_confirmed_df
```

```python
us_confirmed_df.to_csv('covid.csv')
```
